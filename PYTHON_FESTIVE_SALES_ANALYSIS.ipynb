```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding ='unicode_escape')
```

```python
df
```

```python
#data cleaning
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True) # dropped null values columns
```

```python
pd.isnull(df).sum() # to check column wis nulls
```

```python
# drop null values in the dataset
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.info()
```

```python
# changing amounts data type from float to integer
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes # to check data type
```

```python
df.columns
```

```python
# rename column
df.rename(columns={'Marital_Status':'Shaadi'}) # only if we use inplace it will change permannetly
```

```python
df.describe()
```

```python
df[['Orders','Amount']].describe() # only checking the required columns
```

EXPLORATORY DATA ANALYSIS

```python
df.columns
```

```python
ax=sns.countplot(x='Gender', data=df)
```

```python
ax=sns.countplot(x='Gender', data=df) # count of customers
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# gender wise total amount spent
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sg = sns.barplot(x = 'Gender', y='Amount', data = sales_gen)
for container in sg.containers:
    sg.bar_label(container, fmt='%d')
```

```python
ax_sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sg = sns.barplot(x = 'Age Group', y='Amount', data = ax_sales)
for container in sg.containers:
    sg.bar_label(container, fmt='%d')
```

```python
ax=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})

bx=sns.barplot(x='State',y='Orders',data=ax)
for container in bx.containers:
    bx.bar_label(container, fmt='%d')
    
```
